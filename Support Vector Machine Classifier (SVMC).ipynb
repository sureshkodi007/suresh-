```python
x = forest.iloc[:,0:28]
y = forest.iloc[:,28] 
```

```python
x
```

```python
y
```

```python
# Normalising the data as there is scale difference
def norm_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
fires = norm_func(x)
fires
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25, stratify = y)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf','linear','poly'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ 
```

```python
gsv.best_score_
```

```python
# Kernal = linear
model_linear = SVC(kernel = "linear")
model_linear.fit(x_train,y_train)
```

```python
pred_linear = model_linear.predict(x_test)
pred_linear
```

```python
np.mean(pred_linear==y_test)
```

```python
acc = accuracy_score(y_test, pred_linear) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, pred_linear)
```

```python
# Kernel = poly
model_poly = SVC(kernel = "poly")
model_poly.fit(x_train,y_train)
```

```python
pred_poly = model_poly.predict(x_test)
pred_poly
```

```python
np.mean(pred_poly==y_test)
```

```python
acc = accuracy_score(y_test, pred_poly) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, pred_poly)
```
