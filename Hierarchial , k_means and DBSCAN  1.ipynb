```python
# K_MEANS CLUSTER 
from sklearn.cluster import KMeans
def norm_func(i):
    x = (i-i.min()) / (i.max() - i.min())
    return (x) 
```

```python
df_norm = norm_func(data.iloc[:,1:])  
df_norm.head(10) 
```

```python
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

```python
clf = KMeans(n_clusters=4)
y_kmeans = clf.fit_predict(df_norm)  
y_kmeans
clf.labels_
```

```python
y_kmeans
```

```python
clf.cluster_centers_ 
```

```python
clf.inertia_
```

```python
# DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
from sklearn.cluster import DBSCAN 
```

```python
data2 = data.drop(['Unnamed: 0'],axis=1)  
data2
```

```python
array=data2.values
array 
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array) 
X   
```

```python
dbscan = DBSCAN(eps=2, min_samples=5) 
dbscan.fit(X)                      
```

```python
dbscan.labels_
```

```python
data3=pd.DataFrame(dbscan.labels_,columns=['cluster']) 
pd.set_option("display.max_rows", None)
```

```python
data3
```

```python
df1 = pd.concat([data,data3],axis=1)  
df1
```

```python
df1.head()
```

```python
plt.style.use('classic')
df1.plot(y ="Murder",x ="Rape",c=dbscan.labels_ ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using DBScan')
```

```python
plt.style.use('classic')
df1.plot(y ="UrbanPop",x ="Rape",c=dbscan.labels_ ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using DBScan')
```
