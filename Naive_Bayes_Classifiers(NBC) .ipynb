```python
# Fit the naive_bayes model 
# train a Gaussian Naive Bayes classifier on the training set
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
print(cfr(y_test, y_pred))
```

```python
# All classification accuracy_scores
TP = cm[0,0]
TN = cm[1,1]
FP = cm[0,1]
FN = cm[1,0]
```

```python
print(TP,TN,FP,FN)
```

```python
# Classification accuracy_score
classification_accuracy = (TP + TN) / float(TP + TN + FP + FN)
classification_accuracy.round(3)
```

```python
# Classification error_score
# print classification error
classification_error = (FP + FN) / float(TP + TN + FP + FN)
classification_error.round(3)
```

```python
# Precission_score
precision = TP / float(TP + FP)
precision.round(3)
```

```python
# Recall_score
recall = TP / float(TP + FN)
recall.round(3)
```

```python
# Specificity_score
specificity = TN / (TN + FP)
specificity.round(3)
```

```python
# Histogram plot
# adjust the font size 
plt.rcParams['font.size'] = 13
plt.hist(y_pred1, bins = 10)
# set the title of predicted probabilities
plt.title('Predicted probabilities of salaries >50K')
plt.xlim(0,1) # x_axis limits 
# Tittle 
plt.xlabel('Predicted probabilities of salaries >50K')
plt.ylabel('Frequency')
```

```python
# ROC(Reciever Operating Characteristics) - AUC(Area Under the Curve) Curves and Accuracy_Scores
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred1, pos_label = '>50K')
print(fpr, tpr, thresholds)
```

```python
plt.figure(figsize=(10,5))
plt.plot(fpr, tpr, linewidth=2)
plt.plot([0,1], [0,1], 'k--' )
plt.rcParams['font.size'] = 13
plt.title('ROC curve for Gaussian Naive Bayes Classifier for Predicting Salaries')
plt.xlabel('FP_Rate (1 - Specificity)')
plt.ylabel('TP_Rate (Sensitivity)')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
ROC_AUC = roc_auc_score(y_test, y_pred1)
ROC_AUC.round(3)
```
